```
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline

import os
path='/content/drive/My Drive/PDI/'
os.chdir(path)
```

```
def display4(a,b,c,d, title1 = 'Original', title2= 'Editado', title3= 'Editado', title4= 'Editado'):
    plt.figure(figsize=(14,14))
    plt.subplot(221), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.imshow(b,cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(223), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(224), plt.imshow(d,cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

**BGR COLOR SPACE**


```
image_BGR= cv2.imread('rub07.jpg',cv2.IMREAD_COLOR)
#b,g,r = cv2.split(img_BGR)

b= image_BGR[:,:,0]
g= image_BGR[:,:,1]
r= image_BGR[:,:,2]

display4(image_BGR,b,g,r,'BGR','b','g','r')
```

```
count = 2
plt.figure(figsize=(12,4))
plt.subplot(1,4,1), plt.imshow(image_BGR/255.0), plt.title("original BGR")
plt.axis("off")
for i, lab in enumerate(["B","G","R"]):
  crgb = np.zeros(image_BGR.shape)
  crgb[:,:,i] = image_BGR[:,:,i]
  plt.subplot(1,4,count), plt.imshow(crgb/255.0), plt.title(lab)
  plt.axis("off")
  count +=1
  
plt.show()
```

```
image_RGB=cv2.cvtColor(image_BGR,cv2.COLOR_BGR2RGB)
r,g,b = cv2.split(image_RGB)

display4(image_RGB,r,g,b,"RGB","r","g","b")
```

```
count = 2
plt.figure(figsize=(12,4))
plt.subplot(1,4,1), plt.imshow(image_RGB/255.0), plt.title("original BGR")
plt.axis("off")
for i, lab in enumerate(["R","G","B"]):
  crgb = np.zeros(image_RGB.shape)
  crgb[:,:,i] = image_RGB[:,:,i]
  plt.subplot(1,4,count), plt.imshow(crgb/255.0), plt.title(lab)
  plt.axis("off")
  count +=1
  
plt.show()
```
