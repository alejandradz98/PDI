```
def displayGray4(a,b,c,d, title1 = "Original", title2="Edited", title3="Edited"):
  plt.figure(figsize=(20,20))
  plt.subplot(141), plt.imshow(a,cmap='gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(142), plt.imshow(b,cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(143), plt.imshow(c,cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.subplot(144), plt.imshow(d,cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.show()
  

def displayGray3(a,b,c,title1 = "Original", title2="Edited", title3="Edited"):
  plt.figure(figsize=(15,15))
  plt.subplot(131), plt.imshow(a,cmap='gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(132), plt.imshow(b,cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(133), plt.imshow(c,cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def displayGray2(a,b,title1 = "Original", title2="Edited"):
  plt.figure(figsize=(10,10))
  plt.subplot(121), plt.imshow(a,cmap='gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(b,cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def displayColor3(a,b,c,title1 = "Original", title2="Edited", title3="Edited"):
  plt.figure(figsize=(10,10))
  plt.subplot(131), plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(132), plt.imshow(b), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(133), plt.imshow(c), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def displayColor2(a,b,title1 = "Original", title2="Edited"):
  plt.figure(figsize=(10,10))
  plt.subplot(121), plt.imshow(a), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(b), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def readRGB(imgPath):
  img=cv2.imread(imgPath)
  img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  return img
```

**FILTROMEDIANO**

```
img= cv2.imread("saltlgle.png",0)
median= cv2.medianBlur(img,3)
median= cv2.medianBlur(median,3) 
median = cv2.medianBlur(median,3) 


displayGray2(img,median)
```

```
img= cv2.imread("saltpepper.png",0)
median= cv2.medianBlur(img,5)
median= cv2.medianBlur(median,5) 
median1 = cv2.medianBlur(median,7) 
median1 = cv2.medianBlur(median,7) 
median1 = cv2.medianBlur(median,7) 

displayGray3(img,median,median1)
```

**FILTROS DERIVADOS DE REALCE(SHARPENNING)**

```
img=cv2.imread("saltlgle.png",0)

median= cv2.medianBlur(img,3)
median= cv2.medianBlur(median,3) 
median = cv2.medianBlur(median,3) 

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1]])
sharpened=cv2.filter2D(median, -1, kernel_sharpening)
displayGray3(img,sharpened,median)

img= cv2.imread("saltpepper.png",0)
median= cv2.medianBlur(img,5)
median= cv2.medianBlur(median,5) 
median1 = cv2.medianBlur(median,7) 
median1 = cv2.medianBlur(median,7) 
median1 = cv2.medianBlur(median,7) 

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1]])
sharpened=cv2.filter2D(median, -1, kernel_sharpening)


displayGray3(img,median1,sharpened)
```

**FILTRO BILATERAL**

```
img= cv2.imread("saltlgle.png",0)

#gaussianBlur

median= cv2.medianBlur(img,3)
median= cv2.medianBlur(median,3) 
median = cv2.medianBlur(median,3) 

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1]])
sharpened=cv2.filter2D(median, -1, kernel_sharpening)
bilateral= cv2.bilateralFilter(sharpened,9,75,75)
displayGray4(img,sharpened,median,bilateral)


```

**LAPLACIANO**

```
img=cv2.imread("furi.jpg",0)
lapl= cv2.Laplacian(img,cv2.CV_64F, ksize=5)
displayGray2(img,lapl)
```

```
img=cv2.imread("furi.jpg",0)
kernel_sharpening = np.array([[0,1,0],
                              [1,-4,1],
                              [0,1,0]])
sharpened=cv2.filter2D(img, -1, kernel_sharpening)

kernel_sharpening11= np.array([[0,-1,0],
                              [-1,4,-1],
                              [0,-1,0]])
sharpened11=cv2.filter2D(img, -1, kernel_sharpening11)

kernel_sharpening1 = np.array([[-1,-1,-1],
                              [-1,8,-1],
                              [-1,-1,-1]])
sharpened1=cv2.filter2D(img, -1, kernel_sharpening1)

kernel_sharpening2 = np.array([[1,1,1],
                              [1,-8,1],
                              [1,1,1]])
sharpened2=cv2.filter2D(img, -1, kernel_sharpening2)

displayGray2(img,sharpened)
displayGray2(img,sharpened11)
displayGray2(img,sharpened1)
displayGray2(img,sharpened2)
```

**FILTROS DERIVATIVOS en X y Y**

```
img = cv2.imread('furi2.jpg',0)
x_sobel = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)

img = cv2.imread('furi2.jpg',0)
x_sobel1 = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

displayGray2(img,x_sobel,"original","derivada x")
displayGray2(img,x_sobel1,"original","derivada y")
```

```
img=cv2.imread("furi2.jpg",0)
kernel_sharpening = np.array([[-1,0,1],
                              [-2,0,2],
                              [-1,0,1]])
sharpened=cv2.filter2D(img, -1, kernel_sharpening)

kernel_sharpening11= np.array([[-1,-2,-1],
                              [0,0,0],
                              [1,2,1]])
sharpened11=cv2.filter2D(img, -1, kernel_sharpening11)

displayGray2(img,sharpened)
displayGray2(img,sharpened11)
```
